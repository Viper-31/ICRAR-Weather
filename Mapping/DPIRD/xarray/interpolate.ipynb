```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt

# ---------------- CONFIG ----------------
nc_file = "DPIRD_stations_combined_refined.nc"
station_name = "Allanooka"
var_name = "airTemperature"

# Choose a specific day to inspect the interpolation closely
test_date = "2024-01-01" 
# ----------------------------------------

# 1. Load the dataset
ds = xr.open_dataset(nc_file)
```

```python
# 2. Extract the data for the specific station and convert to a Series
# We drop NaNs to ensure the interpolator has valid coordinates
ts_original = ds[var_name].sel(station=station_name).to_series()

# 3. Slice a small window for testing (easier to see the curves)
ts_test = ts_original.loc[test_date]

# 4. Perform Interpolation
# Resample to 5-minute ('5min' or '5T')
# We use PCHIP via pandas (requires scipy)
ts_5min_pchip = ts_test.resample("5min").interpolate(method='pchip')
ts_5min_linear = ts_test.resample("5min").interpolate(method='linear')
```

```python
# 5. Plotting the comparison
plt.figure(figsize=(12, 6))

# The 5-min Interpolated Lines
plt.plot(ts_5min_pchip.index, ts_5min_pchip.values, 
         label='5-min PCHIP (Smooth)', color='black', linewidth=1.5, zorder=2)

plt.plot(ts_5min_linear.index, ts_5min_linear.values, 
         label='5-min Linear (Pointy)', color='orange', alpha=0.7)

# The Original 15-min Data Points
plt.scatter(ts_test.index, ts_test.values, 
            color='red', label='Original 15-min Data', s=30, zorder=3)

# # Interpolated Data Points (for clarity)
# plt.scatter(ts_5min_pchip.index, ts_5min_pchip.values, 
#             color='black', s=10, alpha=0.5, label='PCHIP Interpolated Points', zorder=1)

plt.title(f"Interpolation Test: {station_name} - {test_date}")
plt.ylabel("Temperature (°C)")
plt.xlabel("Time")
plt.legend()
plt.grid(True, alpha=0.3)

# Zoom in on a 4-hour window to see the "curve" clearly
start_zoom = pd.Timestamp(test_date + " 08:00:00")
end_zoom = pd.Timestamp(test_date + " 12:00:00")
plt.xlim(start_zoom, end_zoom)

plt.show()

print("Interpolation test complete. No changes were saved to the .nc file.")
```
